```python
import pandas as pd
```

```python
df = pd.read_csv("2005_demo.csv")
df2 = pd.read_csv("2005_coffee.csv")
df=df[df["SEQN"]>1]
df2=df2[df2["SEQN"]>1]
```

```python
df_col = df[['SEQN',"RIDAGEYR","RIDRETH1","DMDMARTL","DMDFMSIZ","DMDEDUC3","DMDEDUC2","INDHHINC","SIALANG"]]
```

```python
df_col = df_col.rename(columns={'SEQN':"ID","RIDAGEYR":"AGE","RIDRETH1":"Ethnicity","DMDMARTL":"Married","DMDFMSIZ":"Family Size","DMDEDUC3":"Education1","DMDEDUC2":"Education2","INDHHINC":"Income","SIALANG":"Language"})
```

```python
"ID","AGE","Ethnicity","Married","Family Size","Education 1","Education 2","Income","Language"
```

```python
{'SEQN':"ID","RIDAGEYR":"AGE","RIDRETH1":"Ethnicity","DMDMARTL":"Married","DMDFMSIZ":"Family Size","DMDEDUC3":"Education1","DMDEDUC2":"Education2","INDHHINC":"Income","SIALANG":"Language"}
```

```python
df_col["Education"] = df_col["Education1"] + df_col["Education2"]
```

```python
df_col.head()
```

```python
cdict = {ID: int(df2.loc[df2['SEQN']==ID]["FFQ0123"]) for ID in df2["SEQN"]}
htdict={ID: int(df2.loc[df2['SEQN']==ID]["FFQ0125"]) for ID in df2["SEQN"]}
itdict={ID: int(df2.loc[df2['SEQN']==ID]["FFQ0124"]) for ID in df2["SEQN"]}
esdict={ID: int(df2.loc[df2['SEQN']==ID]["FFQ0123"]) for ID in df2["SEQN"]}
pkdict = {ID: int(df2.loc[df2['SEQN']==ID]["FFQ0052"]) for ID in df2["SEQN"]}
cbdict = {ID: int(df2.loc[df2['SEQN']==ID]["FFQ0072"]) for ID in df2["SEQN"]}
```

```python
df_col["Coffee"] = df_col["ID"].map(cdict)
df_col["Hot Tea"] = df_col["ID"].map(htdict)
df_col["Iced Tea"] = df_col["ID"].map(itdict)
df_col["Egg Salad"] = df_col["ID"].map(esdict)
df_col["Pickles"] = df_col["ID"].map(pkdict)
df_col["Corned Beef"] = df_col["ID"].map(cbdict)
```

```python
df_col = df_col.dropna(subset =["Coffee"])
```

```python
df_col.to_csv("Processed_Data.csv")
```
